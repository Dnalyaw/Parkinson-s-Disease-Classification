```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits, load_iris
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
from IPython.display import Image
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.losses import SparseCategoricalCrossentropy, BinaryCrossentropy
```

***Goal with this project: try to score the highesst accuracy as possible to classify Parkinson's disease using the data given.***

```python
#Looking over and investigating the data
df = pd.read_csv('parkinsons.data')
df.head()
```

```python
yes = df[df.status==1].drop('name', axis=1)
no =  df[df.status==0].drop('name', axis=1)
```

```python
#notice the significant differences in many of the features
fig = plt.figure(figsize=(18, 30))
for i, column in enumerate(list(yes.columns)):
    fig.add_subplot(8, 3, i + 1)
    sns.histplot(yes[column], kde=True, label='Parkinsons')
    sns.histplot(no[column], kde=True, label='Healthy')
    plt.legend()
```

```python
#Data Preprocessing for Machine Learning and Deep Learning

X = df.drop(['name', 'status'], axis=1) #we don't need these columns to use for later
X.head()
```

```python
min_max_scaler = MinMaxScaler((-1, 1)) #from -1 to 1
std_scaler = StandardScaler() #or standard scalar 
X_scaled = min_max_scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
#Now its easier to see the differences and the data will equally look at each feature
df_scaled.head()
```

```python
y = df['status']
```

```python
#Start modelling
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
y_train.shape, y_test.shape #this is a good amount to test
```

```python
#this function makes it easier to test ML models
def test_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    y_prediction = model.predict(X_test)
    sns.heatmap(confusion_matrix(y_test, y_prediction), annot=True)
    plt.ylabel("True Values")
    plt.xlabel("Predicted Values")
    print("Score: " + str(score))
    return score
```

```python
#LogisticRegression
LR_score = test_model(LogisticRegression(), X_train, X_test, y_train, y_test)
```

```python
#Support Vector Machines
test_model(SVC(), X_train, X_test, y_train, y_test)
```

```python
#Random Forest Algorithm
test_model(RandomForestClassifier(), X_train, X_test, y_train, y_test)
```

```python
from xgboost import XGBClassifier
```

```python
#XGBClassifier seems to be the most accurate in the model predictions
XGB_score = test_model(XGBClassifier(), X_train, X_test, y_train, y_test)
```

```python
#Neural Network test
NN_model = Sequential(
    [
        Input(shape=(X.shape[1],)), #might add regularization if overfitting occurs
        Dense(45, activation='relu', name = 'L1', kernel_regularizer=tf.keras.regularizers.L2(0.01)), #regularization did help it
        Dense(20, activation='relu', name = 'L2'),
        Dense(10, activation='relu', name = 'L3'),
        Dense(1, activation='sigmoid')
    ]
)
```

```python
NN_model.summary()
```

```python
NN_model.compile(loss=BinaryCrossentropy(), #Binary Crossentropy is for binary classification
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)) #Adam Optimizer
NN_model.fit(X_train, y_train, epochs=1000) #100 epochs did 80%, the apex was around 1000 epochs which helped significantly
```

```python
#make prediction on test set
NN_y_pred = NN_model.predict(X_test)
```

```python
#turn it into actual predictions rather than probabilities
NN_pred = (NN_y_pred>0.5).astype('int32')
NN_score = accuracy_score(NN_pred, y_test)
```

```python
#the softmax algorithm is known to work better in calculations, let's try out if it will improve the neural network

softmax_model = Sequential(
    [
        Input(shape=(X.shape[1],)), #might add regularization if overfitting occurs
        Dense(45, activation='relu', name = 'L1', kernel_regularizer=tf.keras.regularizers.L2(0.01)),
        Dense(20, activation='relu', name = 'L2'),
        Dense(10, activation='relu', name = 'L3'),
        Dense(1, activation='linear')
    ]
)
```

```python
softmax_model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
)

softmax_model.fit(
    X_train,y_train,
    epochs=1000
)
```

```python
softmax_prediction = softmax_model.predict(X_test) 
```

```python
#these predictions are calculated in the weights, but need to be converted through the sigmoid function
prediction_p = tf.nn.sigmoid(softmax_prediction)
```

```python
#testing the accuracy of what the data predicted
softmax_final_preds = (prediction_p.numpy() > 0.5).astype('int32')
softmax_score = accuracy_score(softmax_final_preds, y_test)
```

```python
print("1. Neural Network Accuracy: " + str(NN_score*100) + " %")
print("2. Softmax Neural Network Accuracy: " + str(softmax_score*100) + " %")
print("3. XGB Classifier Accuracy: " + str(XGB_score*100) + " %")
print("4. Logistic Regression  Accuracy: " + str(LR_score*100) + " %")
```
